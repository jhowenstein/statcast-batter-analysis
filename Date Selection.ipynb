```python
def return_data_by_index(self,start=None,end=None):
    if start is None:
        start = 0
    if end is None:
        end = len(self.games)
    
    _dfs = []
    for game in self.games[start:end]:
        _dfs.append(game.data)
        
    _df = pd.concat(_dfs)
    
    return _df
```

```python
window = 10
half_window = int(window / 2)

nGames = len(batter.games)
wOBAs = []

for i in range(half_window,nGames-half_window):
    games = batter.games[i-half_window:i+half_window]
    
    print(f'{games[0].date} to {games[-1].date}')
    print(len(games))
    _dfs = []
    atBats = 0
    for game in games:
        _dfs.append(game.data)
        atBats += len(game.atBats)
        
    _df = pd.concat(_dfs)
    print(f'At Bats: {atBats} - Pitches: {_df.shape[0]}')
    
    wOBAs.append(batter.calculate_total_wOBA(df=_df))
```

```python
fig,ax = plt.subplots(figsize=(16,6))

ax.plot(wOBAs)

ax.grid()

ax.set_xlim(0,len(wOBAs))

plt.show()
```

```python
def return_data_by_date(self,start=None,end=None):
    if start is None:
        start = self.games[0].date
    if end is None:
        end = self.games[-1].date
    
    _dfs = []
    for game in self.games:
        start_int = int(start.replace('-',''))
        game_int = int(game.date.replace('-',''))
        end_int = int(end.replace('-',''))
        
        if start_int > game_int:
            continue
            
        if game_int > end_int:
            break
            
        _dfs.append(game.data)
        
    _df = pd.concat(_dfs)
    
    return _df
```
